```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'iris.xls' is in the same directory as your Jupyter Notebook
df = pd.read_excel('iris.xls')

# Create pair plot
sns.pairplot(df, hue='Classification', markers='o')
plt.show()
```

```python
# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='SL', data=df)
plt.title('Box Plot for Sepal Length (SL) by Classification')
plt.show()
```

```python
# Create histograms for each feature
df.hist(figsize=(12, 8), bins=20)
plt.suptitle('Histograms for Each Feature', x=0.5, y=1.02, ha='center', fontsize='large')
plt.show()
```
